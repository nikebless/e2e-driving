```python
from scipy.stats import pearsonr, spearmanr
import pandas as pd

def calculate_pvalues(df, method='pearson'):
    corr_fn = pearsonr if method == 'pearson' else spearmanr
    df = df.dropna()._get_numeric_data()
    dfcols = pd.DataFrame(columns=df.columns)
    pvalues = dfcols.transpose().join(dfcols, how='outer')
    for r in df.columns:
        for c in df.columns:
            pvalues[r][c] = round(corr_fn(df[r], df[c])[1], 4)
    return pvalues
```

```python
columns = ['model_name', 'model_class', 'total_MAE', 'same_track_MAE', 'vista_interventions', 'real_interventions']

# build a dataframe with the results
rows = [
    ['mdn-5-s1', 'mdn', 9.071, None, 8,   15],
    ['mdn-5-s2', 'mdn', 8.941, None,6,   13],
    ['mdn-5-s2', 'mdn', 8.941, None,8,   13],
    ['mdn-5-s3', 'mdn', 8.666, None, 17,  None],
    ['mdn-3-s1', 'mdn', 8.524, None, 19,  None],
    ['mdn-3-s2', 'mdn', 8.927, None, 11,  9],
    ['mdn-3-s3', 'mdn', 9.513, None, 12,  13],
    ['mdn-1-s1', 'mdn', 10.813, None, 55,  100],
    ['mdn-1-s1', 'mdn', 10.813, None, 62,  100],
    ['mdn-1-s2', 'mdn', 12.555, None, 112, None],
    ['mdn-1-s3', 'mdn', 11.891, None, 84,  None],
]

df = pd.DataFrame(rows, columns=columns)
```
